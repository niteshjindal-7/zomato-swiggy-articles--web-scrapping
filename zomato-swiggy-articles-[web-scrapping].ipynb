```python
# Importing the libraries
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
from pandas import DataFrame
from datetime import datetime

import csv
import time
import os
import numpy as np
import pandas as pd
import re
```

```python
driver = webdriver.Chrome("C:/.../Swiggy-Zomato-Food-Aggregators-Analysis/chromedriver/chromedriver.exe")
driver.get("https://gadgets.ndtv.com/search?searchtext=zomato") 
page = driver.current_url

zomato_pages=[]
elems = driver.find_elements_by_xpath("//div[@class='caption_box']//a[@href]")
for elem in elems:
    zomato_pages.append(elem.get_attribute("href"))
page2 = driver.find_element_by_xpath("//a[@class='{page:2}']")
page2.location_once_scrolled_into_view
page2.click()
time.sleep(10)
elems2 = driver.find_elements_by_xpath("//div[@class='caption_box']//a[@href]")
for elem2 in elems2:
    zomato_pages.append(elem2.get_attribute("href"))
page3 = driver.find_element_by_xpath("//a[@class='{page:3}']")
page3.location_once_scrolled_into_view
page3.click()
time.sleep(10)
elems3 = driver.find_elements_by_xpath("//div[@class='caption_box']//a[@href]")
for elem3 in elems3:
    zomato_pages.append(elem3.get_attribute("href"))
page4 = driver.find_element_by_xpath("//a[@class='{page:4}']")
page4.location_once_scrolled_into_view
page4.click()
time.sleep(10)
elems4 = driver.find_elements_by_xpath("//div[@class='caption_box']//a[@href]")
for elem4 in elems4:
    zomato_pages.append(elem4.get_attribute("href"))

print(zomato_pages)
```

```python
# Creates a new .csv file that the scraped data will be written to
csv_file = open('"C:/.../Swiggy-Zomato-Food-Aggregators-Analysis/articles_data/Zomato_Articles.csv', 'w', encoding="utf-8-sig", newline="")

# Defining Writer and Column values for the CSV file
writer = csv.writer(csv_file)
writer.writerow(
    ['Article_title', 'Article_date','Article_content'])
```

After the search result are displayed, scrape the name, cuisines, cost, timings and collections of each restaurant on the page.
Now navigate to each of the restaurant on the searched page and scrape the information.
For each restaurant collect data of reviewer, review text, reviewer metadata (which includes number of reviews, number of followers and number of photos) and review time.

```python
start_time = datetime.now()

#Outer for loop iterates through all the restaurant pages and scrapes the data into csv file

for page in zomato_pages:
    driver.get(page)
    time.sleep(10)
    try:
        print("=" * 40)  # Display the Restaurant page which is being scraped
        present_time = datetime.now()
        print(present_time)
        print("Scraping " + page) 
        ###########################################################
        # Click on All reviews link on the home-page for restaurant 
        ###########################################################
        article_dict = {}
        try:
            title = driver.find_element_by_xpath('//h1').text
        except:
            title = page
        try:
            date= driver.find_element_by_xpath('//span[@class="dtreviewed"]').text
        except:
            date = ""
        try:
            Content = driver.find_element_by_xpath('//div[@class="content_text row description"]').text
        except:
            Content = ""
            date = ""

        article_dict['Article_title'] = title
        article_dict['Article_date'] = date
        article_dict['Article_content'] = Content

        writer.writerow(article_dict.values())
    except Exception as e:
        print(e)
        csv_file.close()
        driver.close()
        break
csv_file.close()
driver.close()
end_time = datetime.now()
print("=" * 60)
```
